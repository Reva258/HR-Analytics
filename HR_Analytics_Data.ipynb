## HR Analytics

```python
### Importing the necessary libraries
import pandas as pd
import numpy as np
```

```python
### Loading the Data set
EmployeeAttrition_data = pd.read_csv("HR-Employee-Attrition.csv", delimiter='	')
```

```python
EmployeeAttrition_data
```

```python
EmployeeAttrition_data.head(10)
```

```python
EmployeeAttrition_data.info()
```

```python
EmployeeAttrition_data.describe()
```

```python
### Checking for null values
EmployeeAttrition_data.isnull().sum()
```

#### The data set does not contains any null values

```python
### Dropping the duplicates in the data set
EmployeeAttrition_data.drop_duplicates()
```

```python
### Dropping redundant columns
columns_to_drop = ['DistanceFromHome', 'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel', 'YearsWithCurrManager', 'DailyRate', 'TrainingTimesLastYear', 'Over18', 'TrainingTimesLastYear']
EmployeeAttrition_data = EmployeeAttrition_data.drop(columns_to_drop, axis=1)
EmployeeAttrition_data
```

```python
### Droping the rows containing any NaN values
EmployeeAttrition_data = EmployeeAttrition_data.dropna()
EmployeeAttrition_data
```

#### The data set did not contain any Nan Values

### Data Visualization

#### Ploting a correlation map for all numeric variables

#### The dataset contains both numeric and categorical variables. So we need to seperate them first

```python
###Seperating categorical variables from numeric variables
numeric_var = EmployeeAttrition_data.select_dtypes(include=['number'])
numeric_var
```

```python
### Correlation map for all numeric variables
Correlation_map = numeric_var.corr()
Correlation_map
```

```python
### Importing the necessary libraries for visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
### Ploting the correlation map using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(Correlation_map, annot=True, cmap='cividis', linewidths=0.5, fmt=".2f")
plt.title("Correlation Map for Numeric Variables for Employee Attrition Dataset")
plt.show()
```

###  Overtime 

```python
labels = EmployeeAttrition_data['OverTime'].value_counts().index
sizes = EmployeeAttrition_data['OverTime'].value_counts().values
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Percentage of Employees Doing Overtime')
plt.show()
```

### Marital Status

```python
category_counts = EmployeeAttrition_data['MaritalStatus'].value_counts()
category_counts
```

```python
maritalstatus_attrition_count = EmployeeAttrition_data.groupby(['MaritalStatus', 'Attrition']).size().unstack(fill_value=0)
```

```python
maritalstatus_attrition_count
```

```python
### Plot a bar graph for the visualization
maritalstatus_attrition_count.plot(kind='barh')
plt.title('Attrition by Marital Status')
plt.xlabel('Count')
plt.ylabel('MaritalStatus')
plt.legend(title='Attrition', loc='upper right')
plt.show()
```

### Attrition by Job Role

```python
plt.figure(figsize=(20, 15))  # Adjust the figure size if needed
sns.countplot(data=EmployeeAttrition_data, x='JobRole', hue='Attrition')
plt.title('Attrition by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.legend(title='Attrition', loc='upper right')
plt.show()
```

### Attrition by Age

```python
sns.boxplot(x='Attrition', y='Age', data=EmployeeAttrition_data)
plt.title('Attrition by Age (Box Plot)')
plt.show()
```

### Attrition by Gender 

```python
### Counting the occurrences of each combination using groupby and count
gender_attrition_count = EmployeeAttrition_data.groupby(['Gender', 'Attrition']).size().unstack(fill_value=0)
```

```python
gender_attrition_count
```

```python
### Plot a bar graph for the visualization
gender_attrition_count.plot(kind='bar')
plt.title('Attrition by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Attrition', loc='upper right')
plt.show()
```

### Attrition by Education field 

```python
### Counting the occurrences of each combination using groupby and count
educationfield_attrition_count = EmployeeAttrition_data.groupby(['EducationField', 'Attrition']).size().unstack(fill_value=0)
```

```python
educationfield_attrition_count
```

```python
### Plot a bar graph for the visualization
educationfield_attrition_count.plot(kind='bar')
plt.title('Attrition by EducationField')
plt.xlabel('EducationField')
plt.ylabel('Count')
plt.legend(title='Attrition', loc='upper right')
plt.show()
```

### Attrition by Department 


```python
### Counting the occurrences of each combination using groupby and count
dept_attrition_count = EmployeeAttrition_data.groupby(['Department', 'Attrition']).size().unstack(fill_value=0)
dept_attrition_count
```

```python
### Plot a bar graph for the visualization
dept_attrition_count.plot(kind='barh')
plt.title('Attrition by Department')
plt.xlabel('Count')
plt.ylabel('Department')
plt.legend(title='Attrition', loc='upper right')
plt.show()
```

### Relation between Overtime and Age

```python
sns.boxplot(x='OverTime', y='Age', data=EmployeeAttrition_data)
plt.title('Age Distribution by Overtime')
plt.show()
```

## Interpretation :
1. According to the analysis we noticed that employees in the age group between 15 to 26 years are leaving the company at a higher rate as compared to others.
2. The Research and Development Department has the highest attrition.
3. Also, the employees having Life Sciences background are leaving the company at higher rate.
4. Laboratory Technician followed by Sales Executive are the job roles showing higher attrition.
5. More of male employees are leaving company rather than female employees.
